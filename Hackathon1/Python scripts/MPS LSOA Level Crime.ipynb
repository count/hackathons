## MPS LSOA Level Crime (most recent 24 months)

```python
# Usual imports that are needed for most analysis
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from datetime import datetime
df = pd.read_csv('MPS LSOA Level Crime (most recent 24 months).csv')
df.describe()
```

```python
df.head()
```

```python
t = pd.melt(df, id_vars = ["LSOA Code", "Borough","Major Category", "Minor Category"],var_name="Date", value_name="Number of reports")
```

```python
t.head()
```

```python
pd.to_datetime(t['Date'],format = '%Y%m').apply(lambda x: x.strftime('%m-%y'))
```

```python
t.describe()
```

```python
t.to_csv('MPS LSOA Level Crime (most recent 24 months) (melted).csv', sep='\t', index=False)
```
