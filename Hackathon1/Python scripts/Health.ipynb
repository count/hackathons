```python
# Usual imports that are needed for most analysis
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from datetime import datetime
import io
import requests
```

```python
# In order to find an indicator we can use the dataset metadata
metadata = pd.read_csv('LocalAuthorityHealthProfiles.metadata.csv')
metadata[metadata['Indicator'].str.contains("cancer")]
```

```python
df = pd.read_csv('LocalAuthorityHealthProfiles-CountyUA.data.csv')
df.describe()
```

```python
t = df[~df['Area Name'].str.contains("England")]
t = t[~t['Area Name'].str.contains("region")]
```

```python
t['Time period Sortable'] = pd.to_datetime(t['Time period Sortable'], format='%Y0000')
t['Time period Sortable'] = t['Time period Sortable'].dt.strftime('%Y')
```

```python
t = t.rename(columns={'Time period Sortable': 'Year', 'Area Name': 'Borough'})
```
