Now, let's see how the median housing price has changed over time for each borough. 


```python
count = CountAPI()
count.set_api_token(token)
table = count.upload(data = df,name = 'Housing Over Time')
visual = table.upload_visual(x = 'Year',y = 'Median House Price (GBP)',aggregates = {'Median House Price (GBP)':'avg'},
                            color = 'Area name')
visual.embed()
```

With the scale, it's hard to see which boroughs are actually increasing the most. Let's find the Year Over Year Percent Increase. 

```python
yoy = df[['Area name','Year','Median House Price (GBP)']]
yoy = yoy[yoy['Median House Price (GBP)'].isnull()==False]
yoy_piv = yoy.pivot_table(index = 'Year',columns = 'Area name',values = 'Median House Price (GBP)')
```

```python
yoy_piv.head()
```

The data must be in this format to pd.pct_change() to work. Check the documentation here: 
https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.pct_change.html

```python
changes = yoy_piv.pct_change().reset_index()
changes.head()
```

```python
changes['avg_change']=changes.iloc[:,1::].mean(axis = 1)
```

Now let's pivot it back to the way we had it before...

```python
changes_melt = pd.melt(changes, id_vars ='Year',value_vars = changes.columns[1:-1],value_name = 'YoY Change',var_name = 'Area name')
```

```python
changes_melt.head()
```

We can add a few columns to see how the percent change compares to the average change across all Boroughs.

```python
changes_melt = changes_melt.merge(changes[['Year','avg_change']],left_on = 'Year',right_on = 'Year',how = 'left')
changes_melt['above_avg']=np.where(changes_melt['YoY Change']>changes_melt['avg_change'],'above','below')
```

```python
changes_melt.head()
```

```python
changes_table = count.upload(data = changes_melt,name = 'Changes by Borough')
visual = changes_table.upload_visual(x = 'Year',y = 'YoY Change',color = 'Area name',aggregates = {'YoY Change':'avg'})
visual.embed()
```

Before Kennsington and Chelsea stood out, but now we can see that acutally Newham is seeing housing prices increasing the most. 

We can now merge this back with our original data. 

```python
df = df.merge(changes_melt, on = ['Year','Area name'],how = 'left')
```
